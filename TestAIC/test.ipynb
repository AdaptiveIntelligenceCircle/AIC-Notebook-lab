```python
import json
import random 
import math 
from datetime import datetime 

def generate_packet():
    intensity = random.random()  # mức bất thường
    return {
        "timestamp": datetime.now().isoformat(),
        "payload": f"behavior amplitude={round(intensity, 3)}",
        "value": intensity
    }

packets = [generate_packet() for _ in range(20)]
packets
```
